```python
import nemo.collections.nlp as nemo_nlp
import torch

device = 'cuda' if torch.cuda.is_available() else 'cpu'
print(f"Modelo ejecutándose en: {device}")

try:
    model = nemo_nlp.models.TextClassificationModel.from_pretrained(model_name='sentiment_analysis_bert').to(device)
    print("Modelo cargado con éxito.")
except Exception as e:
    print(f"Error al cargar el modelo: {e}")
```

```python
# Prueba de inferencia
frases_test = ["La formación en IA es apasionante", "El sistema ha fallado inesperadamente"]
predicciones = model.classifytext(queries=frases_test, batch_size=2)

for f, p in zip(frases_test, predicciones):
    print(f"Frase: {f} | Sentimiento: {p}")
```
